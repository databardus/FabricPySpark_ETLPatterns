# This notebook showcases a basic ETL pattern.
## Does it cover many of the variances found in ETL processes? No, but it's a place to start.

```python
#Value populated here, but as a parameter, this could be passed in
#As relative paths, you'll need to connect a default lakehouse to this notebook.
file_source = 'Files/Input/updated_music_industry_data.csv'
delta_destination = "Tables/ETLFinal/music_industry_data"
```

# Extract

```python
#Build Extract
#Include row filtering

df = spark.read.format("csv").option("header","true").load(file_source).limit(5000)
# df now is a Spark DataFrame containing CSV data from "Files/Input/updated_music_industry_data.csv".
display(df)
```

# Transform

```python
#Lookup surrogate key
#Clean up duplicates
#Clean up target field
#Try using reusable function
```

```python
# Code generated by Data Wrangler for PySpark DataFrame

def clean_data(df):
    # Drop duplicate rows in columns: 'Name', 'FirstName' and 7 other columns
    df = df.dropDuplicates(['Name', 'FirstName', 'LastName', 'Age', 'Genre', 'Address', 'Email', 'PhoneNumber', 'AgeBucket'])
    # Drop column: 'Name'
    df = df.drop('Name')
    return df

df_clean = clean_data(df)
display(df_clean)
```

# Load

```python
#Write data to Delta table in append mode, unless you can't do anything other than a full load
#Keep it simple

df_clean.write.mode("overwrite").format("delta").save(delta_destination)
```

```python
#Display the end results of the ETL process above by querying from the delta table.
df = spark.sql("SELECT * FROM LHS_Databard_Demo.ETLFinal.music_industry_data LIMIT 1000")
display(df)
```
